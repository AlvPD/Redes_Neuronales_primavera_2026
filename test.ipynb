```python
# - read the input data:

import mnist_loader
training_data, validation_data, test_data = mnist_loader.load_data_wrapper()
training_data = list(training_data)

# - network.py example:
import network

net = network.Network([784, 30, 10])
net.SGD(training_data, 30, 10, 3.0, test_data=test_data)
```

```python
# Guardamos los datos
import pickle

archivo = open("red_prueba.pkl",'wb')
pickle.dump(net,archivo)
archivo.close()

archivo_lectura = open("red_prueba.pkl",'rb')
net = pickle.load(archivo_lectura)
archivo_lectura.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mnist_loader

# 1. Recargamos los datos para "resetear" el iterador y lo convertimos a lista

_, _, test_data_iter = mnist_loader.load_data_wrapper()
test_data_list = list(test_data_iter)

# 2. Seleccionamos un índice fijo
indice = 27
imagen_vector, etiqueta_real = test_data_list[indice]

# 3. Preparar la imagen (reshape de vector a matriz 28x28)
imagen_2d = imagen_vector.reshape(28, 28)

# 4. Inferencia: Pasar el vector por la red
activaciones = net.feedforward(imagen_vector)
prediccion = np.argmax(activaciones)

# 5. Visualización
plt.figure(figsize=(4, 4))
plt.imshow(imagen_2d, cmap='gray')
plt.title(f"Etiqueta Real: {etiqueta_real} | Predicción: {prediccion}")
plt.axis('off')
plt.show()

print(f"La red neuronal dice que este número es un: {prediccion}")
```
